# KMP

```python
global SR_NO
SR_NO=0  

from IPython.core.debugger import set_trace
import sys

def KMPSearch(txt,pat): 
    global SR_NO 
    M = len(pat) 
    N = len(txt) 
    
    if(M>N):
        raise Exception("Pattern Length Can not be more than Text length")

   
    lps = [0]*M 
    j = 0 
  
   
    computeLPSArray(pat, M, lps) 
  
    i = 0 
    count = 0 
    count_str=''
    occur=''
    flag = 0 
    while i < N: 
        if pat[j] == txt[i]: 
            count+=1
            i += 1
            j += 1
  
        if j == M: 
            #print("Found pattern at index " + str(i-j) + " and Number Of Comparisions " + str(count))
            occur=occur+str(i-j)+','
            count_str=count_str+str(count)+','    
            flag = 1
            
            j = lps[j-1] 
  
        # mismatch after j matches 
        elif i < N and pat[j] != txt[i]: 
            count+=1
            # Do not match lps[0..lps[j-1]] characters, 
            # they will match anyway 
            if j != 0: 
                j = lps[j-1] 
            else: 
                i += 1
        
    
    SR_NO=SR_NO+1 
    if flag==0:
        count_str=str(count)
        print(SR_NO,' '*(len('No. ')-len(str(SR_NO))),"No match ",' '*(len('Pattern occur at Index')+7-len("No matc")),count_str)
    else:
        print(SR_NO,' '*(len('No. ')-len(str(SR_NO))),occur[:-1],' '*(len('Pattern occur at Index')+7-len(occur[:-3])),count_str[:-1])
        

def computeLPSArray(pat, M, lps): 
    len = 0
  
    lps[0] 
    i = 1
    
    while i < M: 
        if pat[i]== pat[len]: 
            len += 1
            lps[i] = len
            i += 1
        else: 
           
            if len != 0: 
                len = lps[len-1] 
  
               
            else: 
                lps[i] = 0
                i += 1

def main():

	print("No.   Pattern occur at Index        No. Comparisons Required")
	txt = "FINDINAHAYSTACKNEEDLEIN"
	pat = "NEEDLE"
	KMPSearch(txt, pat)
    
	txt = "BARD LOVED BANANAS"
	pat = "BAOBAB"
	KMPSearch(txt, pat)

	txt = "AABAACAADAABAABA"
	pat = "AABA"
	KMPSearch(txt, pat)

	txt = "AAAAAAAAAAAAAAAAAA"
	pat = "AAAAA"
	KMPSearch(txt, pat)


	txt = "Happy tHANKS-Giving"
	pat = "Giv"
	KMPSearch(txt, pat)


	txt = "This is A SiMple Example which IS SiMple"
	pat = "SiMple"
	KMPSearch(txt, pat)


	txt = "WHERE IS STRING ABCD IN THIS SORRY ABCDE"
	pat = "ABCDE"
	KMPSearch(txt, pat)

	txt = "%#%$$$---$$++@%%--$$"
	pat = "%"
	KMPSearch(txt, pat)

	txt = "AAAAAAAAAAAAAAAAAB"
	pat = "AAAA"
	KMPSearch(txt, pat)
    
	txt = "ABABABCABABABCABABABC"
	pat = "ABABAC"
	KMPSearch(txt, pat)

    
	txt = "JIM_SAW_ME_IN_A_BARBERSHOP"
	pat = "BARBER"
	KMPSearch(txt, pat)

if __name__ == '__main__': 
	main() 


```

# Boyer-Moore-Horspool

```python
NO_OF_CHARS = 256
global SR_NO
SR_NO=0
def badCharHeuristic(string, size): 


	# Initialize all occurrence as -1 
	badChar = [size]*NO_OF_CHARS 

	# Fill the actual value of last occurrence 
	for i in range(size-1): 
		badChar[ord(string[i])] = size-1-i; 

	# retun initialized list 
	return badChar 

def search(txt, pat): 
	global SR_NO
	m = len(pat) 
	n = len(txt) 
	no_of_comparisons=0
	occur=''
	badChar = badCharHeuristic(pat, m) 
	comparisons=''
	# s is shift of the pattern with respect to text 
	s = 0
	flag=0
    
	while(s <= n-m): 
		j = m-1
 
		while j>=0 and pat[j] == txt[s+j]: 
			no_of_comparisons=no_of_comparisons+1
			j -= 1
            
		if j<0:
			occur=occur+str(s)+','
			comparisons=comparisons+str(no_of_comparisons)+','
			flag=1
            
			s += badChar[ord(txt[s+m-1])] 

		else: 
			no_of_comparisons=no_of_comparisons+1
            
			s += badChar[ord(txt[s+m-1])]
	SR_NO=SR_NO+1 
	if flag==0:  
		comparisons=str(no_of_comparisons)
		print(SR_NO,' '*(len('No. ')-len(str(SR_NO))),"No match ",' '*(len('Pattern occur at Index')+7-len("No matc")),comparisons)
	else:
		print(SR_NO,' '*(len('No. ')-len(str(SR_NO))),occur[:-1],' '*(len('Pattern occur at Index')+7-len(occur[:-3])),comparisons[:-1])


# Driver program to test above function 
def main():

	print("No.   Pattern occur at Index        No. Comparisons Required")
	txt = "FINDINAHAYSTACKNEEDLEIN"
	pat = "NEEDLE"
	search(txt, pat)
    
	txt = "BARD LOVED BANANAS"
	pat = "BAOBAB"
	search(txt, pat)

	txt = "AABAACAADAABAABA"
	pat = "AABA"
	search(txt, pat)

	txt = "AAAAAAAAAAAAAAAAAA"
	pat = "AAAAA"
	search(txt, pat)


	txt = "Happy tHANKS-Giving"
	pat = "Giv"
	search(txt, pat)


	txt = "This is A SiMple Example which IS SiMple"
	pat = "SiMple"
	search(txt, pat)


	txt = "WHERE IS STRING ABCD IN THIS SORRY ABCDE"
	pat = "ABCDE"
	search(txt, pat)

	txt = "%#%$$$---$$++@%%--$$"
	pat = "%"
	search(txt, pat)

	txt = "AAAAAAAAAAAAAAAAAB"
	pat = "AAAA"
	search(txt, pat)
    
	txt = "ABABABCABABABCABABABC"
	pat = "ABABAC"
	search(txt, pat)

    
	txt = "JIM_SAW_ME_IN_A_BARBERSHOP"
	pat = "BARBER"
	search(txt, pat)

    
    

if __name__ == '__main__': 
	main() 


```

# Boyer Moore

```python
NO_OF_CHARS = 256
global SR_NO
SR_NO=0  
def badCharHeuristic(string, size): 
    
    badChar = [-1]*NO_OF_CHARS 
  
    
    for i in range(size): 
        badChar[ord(string[i])] = i; 
  
    # retun initialized list 
    return badChar 

def preprocess_strong_suffix(shift, bpos, pat, m): 
  
   
    i = m 
    j = m + 1
    bpos[i] = j 
  
    while i > 0: 
          
      
        while j <= m and pat[i - 1] != pat[j - 1]: 
              
            
            if shift[j] == 0: 
                shift[j] = j - i 
  
            
            j = bpos[j] 
        
        i -= 1
        j -= 1
        bpos[i] = j 
  
# Preprocessing for case 2 
def preprocess_case2(shift, bpos, pat, m): 
    j = bpos[0] 
    for i in range(m + 1): 
          
       
        if shift[i] == 0: 
            shift[i] = j 
              
       
        if i == j: 
            j = bpos[j]

def BMsearch(txt, pat): 
    global SR_NO    
    m = len(pat) 
    n = len(txt) 
    bpos = [0] * (m + 1) 
    flag=0
    occur=''
    shift = [0] * (m + 1) 
  
    
    badChar = badCharHeuristic(pat, m)
    preprocess_strong_suffix(shift, bpos, pat, m) 
    preprocess_case2(shift, bpos, pat, m)
  
    
    s = 0
    s1 = 0
    s2 = 0
    count_str = ''
    count=0
    
    while(s <= n-m): 
        j = m-1
  
       
        while j>=0 and pat[j] == txt[s+j]: 
            j -= 1
            count+=1
  
       
        if j<0:
            occur=occur+str(s)+','
            count_str=count_str+str(count)+','            
            flag=1
            
            #print("Found Pattern at index = " + str(s) + " and Number Of Comparisions = " + str(count))
  
            s1 = (m-badChar[ord(txt[s+m])] if s+m<n else 1)
            s2 = shift[0]
            if(s1>s2):
                s+=s1
            else:
                s+=s2
        else: 
            count+=1
            
            s1 = max(1, j-badChar[ord(txt[s+j])]) 
            s2 = shift[j + 1] 
            if(s1>s2):
                s+=s1
            else:
                s+=s2
    SR_NO=SR_NO+1 
    if flag==0:
        count_str=str(count)
        print(SR_NO,' '*(len('No. ')-len(str(SR_NO))),"No match ",' '*(len('Pattern occur at Index')+7-len("No matc")),count_str)
    else:
        print(SR_NO,' '*(len('No. ')-len(str(SR_NO))),occur[:-1],' '*(len('Pattern occur at Index')+7-len(occur[:-3])),count_str[:-1])
        
def main():

	print("No.   Pattern occur at Index        No. Comparisons Required")
	txt = "FINDINAHAYSTACKNEEDLEIN"
	pat = "NEEDLE"
	BMsearch(txt, pat)
    
	txt = "BARD LOVED BANANAS"
	pat = "BAOBAB"
	BMsearch(txt, pat)

	txt = "AABAACAADAABAABA"
	pat = "AABA"
	BMsearch(txt, pat)

	txt = "AAAAAAAAAAAAAAAAAA"
	pat = "AAAAA"
	BMsearch(txt, pat)


	txt = "Happy tHANKS-Giving"
	pat = "Giv"
	BMsearch(txt, pat)


	txt = "This is A SiMple Example which IS SiMple"
	pat = "SiMple"
	BMsearch(txt, pat)


	txt = "WHERE IS STRING ABCD IN THIS SORRY ABCDE"
	pat = "ABCDE"
	BMsearch(txt, pat)

	txt = "%#%$$$---$$++@%%--$$"
	pat = "%"
	BMsearch(txt, pat)

	txt = "AAAAAAAAAAAAAAAAAB"
	pat = "AAAA"
	BMsearch(txt, pat)
    
	txt = "ABABABCABABABCABABABC"
	pat = "ABABAC"
	BMsearch(txt, pat)

    
	txt = "JIM_SAW_ME_IN_A_BARBERSHOP"
	pat = "BARBER"
	BMsearch(txt, pat)

if __name__ == '__main__': 
    main() 
```
